## Cambiando la visualización de datos... desde más Matplotlib, al infinito y más allá!

La visualización de datos es una parte cada vez más importante de la ciencia y el análisis de datos. En la clase anterior vimos como Matplotlib es la herramienta _de facto_ para la visualización de datos en Python, pero esto no queda aquí, sino que podemos ir a un nivel superior.

A continuación veremos otras librerías que nos permiten hacer visualizaciones más vistosas, de una manera fácil y accesible.

### Veamos un gráfico default generado por `matplotlib`:

```python
import matplotlib.pyplot as plt
import numpy as np

def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i))
        
sinplot()
```

```python
import seaborn as sns   # magia !

with sns.color_palette("Blues", desat=.7):
    sinplot()
```

## Seaborn

[Seaborn](https://seaborn.pydata.org/) es una increíble biblioteca de visualización para el trazado de gráficos en Python. Proporciona hermosos estilos por defecto y paletas de colores para hacer que los gráficos sean más atractivos. Está construida sobre la biblioteca `matplotlib` y también está estrechamente integrada con las estructuras de datos de `pandas`.

Seaborn tiene como objetivo hacer que la visualización sea la parte central de la exploración y comprensión de los datos. Proporciona APIs orientadas al conjunto de datos, de modo que podemos cambiar entre diferentes representaciones visuales para las mismas variables para una mejor comprensión del conjunto de datos.

### Diferentes categorías de gráficos en Seaborn 

Los gráficos se utilizan básicamente para visualizar la relación entre variables. Esas variables pueden ser completamente numéricas o una categoría como un grupo, clase o división. Seaborn divide los gráficos en las siguientes categorías:

* __Gráficos relacionales__: se utiliza para entender la relación entre dos variables.
* __Gráficos categóricos__: se ocupa para variables categóricas y de cómo se pueden visualizar.
* __Gráficos de distribución__: se utiliza para examinar las distribuciones univariantes y bivariantes.
* __Gráficos de regresión__: pensados principalmente para añadir una guía visual que ayude a enfatizar los patrones en un conjunto de datos durante los análisis exploratorios de datos.
* __Gráficos de matriz__: se utilizan para visualizar la dispersión de los datos.
* __Cuadrículas multigráficos__: útil para dibujar múltiples instancias del mismo gráfico en diferentes subconjuntos del conjunto de datos.

### Uso de Seaborn con Matplotlib

El uso conjunto de Matplotlib y Seaborn es un proceso muy sencillo. Sólo tenemos que invocar la función de trazado de Seaborn de forma normal, y luego podemos utilizar la función de personalización de Matplotlib.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')

# setting the x limit of the plot
plt.xlim(5,7)

plt.show()
```

### Personalización de los gráficos con Seaborn

Seaborn viene con algunos temas personalizados y una interfaz de alto nivel para personalizar el aspecto de los gráficos. Considere el ejemplo anterior donde se utiliza el valor por defecto del Seaborn. Todavía se ve bien y bonito, pero podemos personalizar el gráfico de acuerdo a nuestras propias necesidades. Así que vamos a ver el estilo de los gráficos en detalle.

* Cambiando el estilo: `set_style()` se utiliza para establecer el estilo del gráfico. Esto significa que afecta a cosas como el color de los ejes, si la cuadrícula está activa o no, u otros elementos estéticos. Hay cinco temas disponibles en Seaborn: darkgrid, whitegrid, dark, white, ticks.

* Cambiar el tamaño de la figura: El tamaño de la figura se puede cambiar usando el método `figure()` de Matplotlib. El método crea una nueva figura del tamaño especificado que se pasa en el parámetro `figsize`.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# changing the figure size
plt.figure(figsize = (10, 4))

# Selecting style as white,
# dark, whitegrid, darkgrid
# or ticks
sns.set(style="white")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Removing the spines
sns.despine()

plt.show()
```

* Escalado de los gráficos: Se puede hacer mediante el método `set_context()`. Anula los parámetros por defecto, afectando el tamaño de las etiquetas, líneas y otros elementos del gráfico, pero no al estilo general. El contexto base es "notebook", y los otros contextos son "paper", "talk", y "poster". `font_scale` establece el tamaño de la fuente.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Setting the scale of the plot
sns.set_context("poster")

plt.show()
```

* Fijar el estilo temporalmente: el método `axes_style()` se utiliza para establecer el estilo temporalmente. Se utiliza junto con la sentencia `with`.


```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# Setting the scale of the plot
sns.set_context("paper")

def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

with sns.axes_style('darkgrid'):
    # Adding the subplot
    plt.subplot(211)
    plot()
    
plt.subplot(212)
plot()
```

* Configuración de la paleta de colores por defecto: El método `set_palette()` se utiliza para establecer la paleta de colores por defecto para todos los gráficos. Los argumentos para `color_palette()` y `set_palette()` son los mismos. `set_palette()` cambia los parámetros por defecto de matplotlib.


```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

def plot():
	sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# setting the default color palette
sns.set_palette('vlag')
plt.subplot(211)

# plotting with the color palette
# as vlag
plot()

# setting another default color palette
sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

* `pairplot()` representa la relación entre pares en todo el marco de datos y admite un argumento adicional llamado `hue` para la separación categórica. Lo que hace básicamente es crear un gráfico conjunto entre cada columna numérica posible y tarda un tiempo si el marco de datos es realmente enorme.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.pairplot(data=data, hue='species')
plt.show()
```

* El mapa de calor se define como una representación gráfica de datos que utiliza colores para visualizar el valor de la matriz, se puede trazar utilizando la función `heatmap()`. En ella, para representar los valores más comunes o las actividades más altas se utilizan colores más brillantes, básicamente rojizos, y para representar los valores menos comunes o de actividad se prefieren los colores más oscuros.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("tips")

# correlation between the different parameters
tc = data.corr()

sns.heatmap(tc, annot=True)
plt.show()
```

## Lux

[Lux](https://github.com/lux-org/lux) es una gran herramienta para la visualización y el análisis exploratorio de datos. Es posible tomar cualquier marco de datos y recomendará automáticamente visualizaciones que le ayuden a explorar y compartir los datos. No es necesario que escriba usted mismo ningún código de visualización.

    import lux
    import pandas as pd

A continuación, se puede ver cómo Lux proporciona opciones de visualización para cualquier marco de datos. Todo lo que tienes que hacer es seleccionar el gráfico que quieres, sin necesidad de codificar nada. Esto es un gran ahorro de tiempo, ya que conseguir la sintaxis correcta exacta de un paquete de como Matplotlib o Seaborn puede ser intensivo en tiempo.

![lux](img/lux.gif)
